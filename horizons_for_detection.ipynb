```python
# These ones below are functions that returns 0 when you match the SNR threshold.
# We solve these functions in redshift

def snrsolver(z, rho_threshold, m1, m2,interpo,btype,cosmo):
    
    z=np.abs(z)
    chirp_mass=np.power(m1*m2,3/5.)/np.power(m1+m2,1/5.)
    chirp_mass*=1.99e30
    chirp_mass*=(1+z)
    
    Mtot=m1+m2
    Mtot*=4.9e-6
    Mtot*=(1+z)
    
    lso=1./(Mtot*np.pi*6**1.5)
    distance=cosmo.dl_at_z(z)*3.08e22
    snr=(np.power(constants.G.value*chirp_mass,5./3)*Q)/(6*np.power(constants.c.value,3.)*np.power(np.pi,4/3.)*np.power(distance,2.))

    if (m1+m2)<6.:
        snr*=interp(lso)
    else:
        snr*=interp(2*lso)
    
    snr=np.sqrt(snr)-rho_threshold
    return snr

def snrsolver_2(z, rho_threshold, m1, m2,interpo,btype,cosmo):
    
    z=np.abs(z)
    chirp_mass=np.power(m1*m2,3/5.)/np.power(m1+m2,1/5.)
    chirp_mass*=1.99e30
    chirp_mass*=(1+z)
    
    Mtot=m1+m2
    Mtot*=4.9e-6
    Mtot*=(1+z)
    
    lso=1./(Mtot*np.pi*6**1.5)
    distance=cosmo.dl_at_z(z)*3.08e22
    
    cosiota=np.random.uniform(-1,1,size=len(m1))
    Q=0.2*(np.power(0.5*(1+cosiota**2.),2.)+cosiota**2)
        
    snr=(np.power(constants.G.value*chirp_mass,5./3)*Q)/(6*np.power(constants.c.value,3.)*np.power(np.pi,4/3.)*np.power(distance,2.))
    
    idxBNS=np.where((m1+m2)<=6.)[0]
    snr[idxBNS]=snr[idxBNS]*interp(lso[idxBNS])
    
    idxBNS=np.where((m1+m2)>6.)[0]
    snr[idxBNS]=snr[idxBNS]*interp(2*lso[idxBNS])
    
    snr=np.sqrt(snr)-rho_threshold
    return snr
```

```python
m1,m2=np.linspace(0.5,150,100),np.linspace(0.5,150,100)

zh6=np.zeros_like(m1)
i=0
for mm1,mm2 in zip(m1,m2):
    zh6[i]=fsolve(snrsolver,args=(6,mm1,mm2,interp,'BBH',icarogw.cosmologies.flatLCDM(H0=67.76,Omega_m=0.3097)),xtol=1e-20,x0=0.5)
    i+=1
    
zh9=np.zeros_like(m1)
i=0
for mm1,mm2 in zip(m1,m2):
    zh9[i]=fsolve(snrsolver,args=(9,mm1,mm2,interp,'BBH',icarogw.cosmologies.flatLCDM(H0=67.76,Omega_m=0.3097)),xtol=1e-20,x0=0.5)
    i+=1
    
zh12=np.zeros_like(m1)
i=0
for mm1,mm2 in zip(m1,m2):
    zh12[i]=fsolve(snrsolver,args=(12,mm1,mm2,interp,'BBH',icarogw.cosmologies.flatLCDM(H0=67.76,Omega_m=0.3097)),xtol=1e-20,x0=0.5)
    i+=1
```
